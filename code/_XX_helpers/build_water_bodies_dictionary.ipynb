````python
df = pd.DataFrame(water_bodies, columns=[
    "feature_name","feature_type","aliases","irwm_region","hydrologic_region","notes"
])

def build_all_names(row):
    names = [row["feature_name"]]
    if row["aliases"]:
        names.extend([a.strip() for a in row["aliases"].split("|") if a.strip()])
    seen, out = set(), []
    for n in names:
        if n not in seen: seen.add(n); out.append(n)
    return "|".join(out)

df["all_names"] = df.apply(build_all_names, axis=1)
total_variants = df["all_names"].apply(lambda x: len(x.split("|"))).sum()

print(f"Total water body features: {len(df)}")
print(f"Total name variants: {total_variants}")
print(f"\nBy feature type:")
print(df["feature_type"].value_counts().to_string())

df.to_csv("/sessions/adoring-nifty-feynman/mnt/tijuanabox/int_data/dictionaries/water_bodies_dictionary.csv", index=False)
print("\nSaved to water_bodies_dictionary.csv")
````

## Usage notes for textNet

### Recommended columns for NER lookup
- Use **`all_names`** (pipe-separated) to build your NER dictionary: split on `|` to get all variants
- Filter by `irwm_region` to scope recognition to a specific planning region
- Filter by `hydrologic_region` for DWR's 10-region hydrologic system
- Filter by `feature_type` to focus on specific feature categories

### Notes on geographic scope
- Multi-region features (e.g., San Joaquin River) are pipe-separated in `irwm_region`
- Transboundary features (Colorado River, Truckee River, Tijuana River) are included with California-relevant IRWM regions
- Historically drained features (Tulare Lake, Buena Vista Lake) are included and flagged in `notes`
