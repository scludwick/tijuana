```python
# Add a helper column: all_names (entity_name + aliases) for quick lookup
# This is the union of all name variants — useful for building textNet lookup tables

def build_all_names(row):
    names = [row['entity_name']]
    if row['abbreviation']:
        names.append(row['abbreviation'])
    if row['aliases']:
        names.extend([a.strip() for a in row['aliases'].split('|') if a.strip()])
    # Deduplicate while preserving order
    seen = set()
    unique_names = []
    for n in names:
        if n not in seen:
            seen.add(n)
            unique_names.append(n)
    return '|'.join(unique_names)

df['all_names'] = df.apply(build_all_names, axis=1)

# Count unique name variants total
total_name_variants = df['all_names'].apply(lambda x: len(x.split('|'))).sum()
print(f"Total unique name variants across all entities: {total_name_variants}")
```

## Usage notes for textNet

### Recommended columns for NER lookup
- Use **`all_names`** (pipe-separated) to build your NER dictionary — this is the full set of name variants that should be matched in text.
- Use **`entity_name`** as the canonical label to assign when a match is found.
- Use **`entity_type`** to filter entity classes for specific analyses (e.g., only water agencies).
- Use **`irwm_region`** to restrict NER to entities relevant to a specific plan's region.

### Known disambiguation challenges
- **MID** is used by Modesto, Merced, AND Madera Irrigation Districts — context (region) needed.
- **SCWA** is used by Sonoma County Water Agency, Sacramento County Water Agency, AND Shasta County Water Agency — use full names.
- **RMWD** is used by both Ramona and Rainbow Municipal Water Districts.
- **CWA** refers to Clean Water Action, California Waterfowl Association, AND the federal Clean Water Act.

### Extending the dictionary
- Tribal entities (e.g., Hoopa Valley Tribe, Yurok Tribe, Tule River Tribe) should be added for regions with active tribal involvement in IRWM.
- Local cities and counties frequently appear as co-signatories and stakeholders — consider adding the most active ones by region.
- Resource Conservation Districts (RCDs) are common IRWM partners — a targeted search by region would capture these.
