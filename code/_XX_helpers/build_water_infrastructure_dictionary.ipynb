## Columns
| Column | Description |
|--------|-------------|
| `feature_name` | Canonical name |
| `feature_type` | Dam, Aqueduct/Canal, Pumping Plant, Water Treatment Plant, Desalination Plant, Spreading Ground/Recharge Basin, Flood Control Channel/Structure, Water Conveyance System, Water Recycling Facility, Hydropower Facility |
| `aliases` | Pipe-separated alternate names and common variants |
| `irwm_region` | IRWM planning region(s); pipe-separated for multi-region infrastructure |
| `operator_owner` | Primary operator or owner (DWR, USBR, USACE, MWD, LADWP, SFPUC, etc.) |
| `source` | Pipe-separated source codes; see `output/SOURCES.md` for full URL reference |
| `notes` | Governance notes, capacity, historical significance, ongoing projects |
| `all_names` | Union of feature_name + aliases — **use this column for NER lookup** |

## Usage notes for textNet

### Recommended columns for NER lookup
- Use **`all_names`** (pipe-separated) to build your NER dictionary: split on `|` to get all variants
- Filter by `irwm_region` to scope recognition to a specific planning region
- Filter by `feature_type` to focus on specific infrastructure categories
- Filter by `operator_owner` to focus on a specific agency's infrastructure

### Combining all three dictionaries
To build a comprehensive NER lookup spanning entities, water bodies, and infrastructure:
```r
entity_dict    <- read_csv("/sessions/adoring-nifty-feynman/mnt/tijuanabox/int_data/dictionaries/water_entity_dictionary.csv")      %>% rename(all_names=all_names)

infra_dict     <- read_csv("/sessions/adoring-nifty-feynman/mnt/tijuanabox/int_data/dictionaries/water_infrastructure_dictionary.csv") %>% rename(all_names=all_names)

# Combine and split
combined_lookup <- bind_rows(
  entity_dict %>% select(name=entity_name, all_names, type=entity_type),
  bodies_dict %>% select(name=feature_name, all_names, type=feature_type),
  infra_dict  %>% select(name=feature_name, all_names, type=feature_type)
) %>%
  separate_rows(all_names, sep = "\\|") %>%
  distinct(all_names, .keep_all = TRUE)
```

### Known ambiguities
- **SWP** – State Water Project (system) vs. individual SWP facilities; prefer full names in plan text
- **CVP** – Central Valley Project (system) vs. specific CVP components
- **Dam names vs. reservoir names** – many CA reservoirs share names with their dams (Shasta Dam / Shasta Lake);
  water_bodies_dictionary.csv captures the reservoir; this dictionary captures the dam
- **LADWP infrastructure** – LA Aqueduct (bodies dict) vs. LA Aqueduct Filtration Plant (this dict) — distinct entries
